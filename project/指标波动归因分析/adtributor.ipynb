#### 构造数据

```python
import pandas as pd

# 原始数据
df_uv = pd.DataFrame({
    'channel': ['A', 'B', 'C', 'D', 'ALL'],
    'before_uv': [50000, 10000, 30000, 10000, 100000],
    'after_uv': [100000, 10000, 50000, 25000, 185000]
})

df_pay = pd.DataFrame({
    'channel': ['A', 'B', 'C', 'D', 'ALL'],
    'before_pay': [10000, 1500, 15000, 1000, 27500],
    'after_pay': [20000, 1500, 26000, 1000, 48500]
})

df_cvr = pd.DataFrame({
    'channel': ['A', 'B', 'C', 'D', 'ALL'],
    'before_cvr': [0.2, 0.15, 0.5, 0.1, 0.275],
    'after_cvr': [0.2, 0.15, 0.52, 0.04, 0.262]
})

# 合并
df_merge = df_uv.merge(df_pay, on='channel').merge(df_cvr, on='channel')

# 计算环比增长率
df_merge['uv_change_rate'] = (df_merge['after_uv'] - df_merge['before_uv']) / df_merge['before_uv']
df_merge['pay_change_rate'] = (df_merge['after_pay'] - df_merge['before_pay']) / df_merge['before_pay']
df_merge['cvr_change_rate'] = (df_merge['after_cvr'] - df_merge['before_cvr']) / df_merge['before_cvr']

# 结果展示（可选保留小数位）
df_merge = df_merge.round({
    'uv_change_rate': 4,
    'pay_change_rate': 4,
    'cvr_change_rate': 4
})

df_merge
```

```python
# 只有A渠道发生了变化
# 访客数和购买人数变化一致，因此A渠道转化率不变，但A渠道访客数
# 和购买人数对整体产生结构性的影响，造成整体转化率环比降低
# 9.1%，这9.1%的降幅可以看作A渠道活动前后数据变化对于整体的影
# 响。

import pandas as pd

# 原始数据，仅 A 渠道的 after 数据发生变化，其余渠道保持不变
df_uv = pd.DataFrame({
    'channel': ['A', 'B', 'C', 'D'],
    'before_uv': [50000, 10000, 30000, 10000],
    'after_uv': [100000, 10000, 30000, 10000]  # 只有 A 改变，其它 = before
})

df_pay = pd.DataFrame({
    'channel': ['A', 'B', 'C', 'D'],
    'before_pay': [10000, 1500, 15000, 1000],
    'after_pay': [20000, 1500, 15000, 1000]  # 只有 A 改变，其它 = before
})

df_cvr = pd.DataFrame({
    'channel': ['A', 'B', 'C', 'D'],
    'before_cvr': [0.2, 0.15, 0.5, 0.1],
    'after_cvr': [0.2, 0.15, 0.5, 0.1]  # 只有 A 真实变，其他保持不变（或根据 pay/uv 算也行）
})

# 合并
df_merge = df_uv.merge(df_pay, on='channel').merge(df_cvr, on='channel')

# 添加 ALL 汇总行：汇总其他渠道数据（after 是自动汇总）
row_all = {
    'channel': 'ALL',
    'before_uv': df_merge['before_uv'].sum(),
    'after_uv': df_merge['after_uv'].sum(),
    'before_pay': df_merge['before_pay'].sum(),
    'after_pay': df_merge['after_pay'].sum()
}
row_all['before_cvr'] = row_all['before_pay'] / row_all['before_uv']
row_all['after_cvr'] = row_all['after_pay'] / row_all['after_uv']

# 加入 ALL 行
df_merge = pd.concat([df_merge, pd.DataFrame([row_all])], ignore_index=True)

# 计算环比增长率
df_merge['uv_change_rate'] = (df_merge['after_uv'] - df_merge['before_uv']) / df_merge['before_uv']
df_merge['pay_change_rate'] = (df_merge['after_pay'] - df_merge['before_pay']) / df_merge['before_pay']
df_merge['cvr_change_rate'] = (df_merge['after_cvr'] - df_merge['before_cvr']) / df_merge['before_cvr']

# 保留小数位
df_merge = df_merge.round({
    'before_cvr': 4,
    'after_cvr': 4,
    'uv_change_rate': 4,
    'pay_change_rate': 4,
    'cvr_change_rate': 4
})

df_merge
```

#### 构造一张活动前后数据均不变的表，为后续的控制变量做准备

```python
# df_uv_con = df_uv.copy() 确实是深拷贝（deep copy），在 pandas 中表示：
#  ✅ df_uv_con 和 df_uv 是两个独立的 DataFrame。
#  ✅ 你对 df_uv_con 的修改 不会影响 原始的 df_uv。
#  ✅ 只复制了数据内容本身，而不是引用。
```

```python
df_uv_con = df_uv.copy()
df_uv_con['after_uv'] = df_uv['before_uv']

df_pay_con = df_pay.copy()
df_pay_con['after_pay'] = df_pay['before_pay']
```

#### 控制变量的方式计算A渠道的影响，即其他渠道活动前后数据不变，计算A渠道数据的变化对整体的影响

```python
#所有渠道不变，只有A渠道是真实的值，访客和购买人数均如此
df_uv_con.loc[df_uv_con['channel'] == 'A',:] = df_uv.loc[df_uv['channel'] == 'A',:]
df_pay_con.loc[df_pay_con['channel'] == 'A',:] = df_pay.loc[df_pay['channel'] == 'A',:]
```

```python
df_uv_con
```

#### 重新计算汇总值，并计算A渠道的影响

```python
#对活动前后的访客数进行汇总并重新计算
before_uv_sum = df_uv_con['before_uv'][:-1].sum()
after_uv_sum = df_uv_con['after_uv'][:-1].sum()

#对活动前后的购买人数进行汇总并重新计算
before_pay_sum = df_pay_con['before_pay'][:-1].sum()
after_pay_sum = df_pay_con['after_pay'][:-1].sum()

#计算A渠道对整体的影响，活动后整体转化率和活动前的环比
before_cvr_all = before_pay_sum / before_uv_sum
after_cvr_all = after_pay_sum / after_uv_sum
result_a = (after_cvr_all - before_cvr_all) / before_cvr_all

#显示4位小数的结果，避免小数位太多
print(round(result_a,4))
```

#### 其他渠道也是同样的控制变量逻辑

```python
result_dic = {}

for channel in df_uv['channel'][:-1]:
    
    #构造控制变量的原始表，活动前后数据相等
    df_uv_con = df_uv.copy()
    df_uv_con['after_uv'] = df_uv['before_uv']

    df_pay_con = df_pay.copy()
    df_pay_con['after_pay'] = df_pay['before_pay']
    
    #用索引替换对应渠道的数据
    df_uv_con.loc[df_uv_con['channel'] == channel,:] = df_uv.loc[df_uv['channel'] == channel,:]
    df_pay_con.loc[df_pay_con['channel'] == channel,:] = df_pay.loc[df_pay['channel'] == channel,:]
    
    
    #对活动前后的访客数进行汇总并重新计算
    before_uv_sum = df_uv_con['before_uv'][:-1].sum()
    after_uv_sum = df_uv_con['after_uv'][:-1].sum()

    #对活动前后额购买人数进行汇总并重新计算
    before_pay_sum = df_pay_con['before_pay'][:-1].sum()
    after_pay_sum = df_pay_con['after_pay'][:-1].sum()

    #计算渠道对整体的影响，活动后整体转化率和活动前的环比
    before_cvr_all = before_pay_sum / before_uv_sum
    after_cvr_all = after_pay_sum / after_uv_sum
    result_x = (after_cvr_all - before_cvr_all) / before_cvr_all

    #显示两位小数的结果，避免小数位太多
    result_dic[channel] = result_x

#构造最终的影响结果
result = pd.DataFrame(result_dic,index = ['value']).T

result
```

```python
#单位都是万

data = pd.DataFrame({'维度':['渠道','渠道','渠道','渠道','新老客','新老客','产品','产品','产品','产品','产品','产品'],
                     '元素':['A','B','C','D','新客','老客','C001','C002','C003','C004','C005','C006'],
                     'before':[50,3,18,4,35,40,20,15,10,8,12,10],
                     'after':[78,23,45,4,60,90,40,38,15,20,17,20]})
data
```

### 1.计算惊讶度并排序

#### 计算出每个维度下各元素的p、q的值

```python
len(data['维度'].unique()) 
```

```python
#计算活动前销售额汇总和活动后销售额汇总
#因为有3个大维度，直接汇总是3倍销售额，需要除以维度数量，这里len(data['维度'].unique()等于3
pre_sum = data['before'].sum() / len(data['维度'].unique()) 
aft_sum = data['after'].sum() / len(data['维度'].unique()) 

#计算p和q值
data['p'] = data['before'] / pre_sum
data['q'] = data['after'] / aft_sum
```

```python
import math
#创建一个列表用于收集每个s值
surprises = []

#遍历计算每一行数据的s值
for p,q in zip(data['p'],data['q']):
    #用JS散度公式
    s = 0.5 * (p * math.log10(2 * p / (p + q)) + q * math.log10(2 * q / (p + q)))
    surprises.append(s)

#把计算好的s列表赋值给data
data['surprise'] = surprises

#每个维度下按照s值排序
data.sort_values(['维度','surprise'],ascending = False,inplace = True)

data
```

### 2.计算EP值并筛选元素

#### 计算每一行数据的EP值

```python
#计算出总销售波动，3个维度都在一起，因此也需要除以维度数量
sum_dif = (data['after'].sum() - data['before'].sum()) / len(data['维度'].unique()) 

#计算每一行数据的EP
data['EP'] = (data['after'] - data['before']) / sum_dif

data
```

#### 设定单个元素EP阈值teep和总EP阈值tep，这里分别将teep和tep设置为0.2与0.8

- 根据设定的单个元素EP阈值teep，遍历所有元素的EP值是否高于0.2，如果高于，则通过筛选
- 每一次遍历的同时，把每个维度下通过筛选的元素EP值累加，如果某维度下累加的EP值大于0.8，则该维度停止筛选

```python
#假设单个EP阈值teep = 0.2，总EP阈值tep = 0.8
teep = 0.2
tep = 0.8

#筛选出EP值大于单个EP阈值teep的元素
data_fil = data.loc[data['EP'] >= teep,['维度','元素','surprise','EP']]

#新建一个EP_sum列，即每个维度内EP值做累加，作为和总阈值tep对比的辅助列
data_fil['EP_sum'] = data_fil.groupby('维度')['EP'].cumsum()
data_fil
```

```python
data_fil.loc[data_fil['EP_sum'] >= tep,:]
```

#### 根据总阈值进行筛选
